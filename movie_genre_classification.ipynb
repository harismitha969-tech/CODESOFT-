```python
# MOVIE GENRE CLASSIFICATION - CodeSoft Internship Task

# Import libraries
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Load training dataset
train_data = pd.read_csv(
    '/content/train_data',
    sep=' ::: ',
    engine='python',
    names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION']
)

# Display first 5 rows
print(train_data.head())

# Features and labels
X = train_data['DESCRIPTION']
y = train_data['GENRE']

# Convert text data into TF-IDF features
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=5000
)

X_tfidf = tfidf.fit_transform(X)

# Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.2,
    random_state=42
)

# Create model
model = MultinomialNB()

# Train model
model.fit(X_train, y_train)

# Predict genres
y_pred = model.predict(X_test)

# Accuracy score
accuracy = accuracy_score(y_test, y_pred)

print("\nAccuracy:", accuracy)

# Classification report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, zero_division=0))

# Predict genre for custom movie description
sample = ["A young hero fights dangerous enemies to save the world."]

sample_tfidf = tfidf.transform(sample)

prediction = model.predict(sample_tfidf)

print("\nPredicted Genre:", prediction[0])
```
